```python
import sys
sys.path.append("c:/Users/rljones/Documents/GitHub/ndp/")

import numpy as np
import matplotlib.pyplot as plt
import json
from pprint import pprint
from NDPReduce import ndp

np.set_printoptions(threshold=15)
np.set_printoptions(precision=10)
np.set_printoptions(suppress=True)

ndp_savepath = "c:/Users/rljones/Documents/NDP Data/Binek 10B Cr2O3 Jones 220623/"
```

```python
# Create instance for this sample
ron = ndp()
ron.runschema()
#ron.saveAtoms(ndp_savepath, 'binek.csv')
```

```python
pprint(ron.data)
```

```python
fig, axs = plt.subplots(3, 1, sharex=True)

fig.suptitle('Raw Counts')

# Plot each graph, and manually set the y tick values
axs[0].set_ylabel('Sample')
axs[0].plot(ron.detector["Energy"], ron.data["Sam Dat"]["Counts"])
axs[0].set_yscale('log')

axs[1].set_ylabel('Background')
axs[1].plot(ron.detector["Energy"], ron.data["Bgd Dat"]["Counts"])
axs[1].set_yscale('log')

axs[2].set_ylabel('Reference')
axs[2].plot(ron.detector["Energy"], ron.data["Ref Dat"]["Counts"])
axs[2].set_yscale('log')

plt.show()
```

```python
fig, axs = plt.subplots(5, 1, sharex=True)

fig.suptitle('Sample Progression')

# Plot each graph, and manually set the y tick values
axs[0].set_ylabel('Raw')
axs[0].plot(ron.detector["Corr Depth"], ron.data["Sam Dat"]["Counts"])
axs[0].set_yscale('linear')

axs[1].set_ylabel('Deadtime')
axs[1].plot(ron.detector["Corr Depth"], ron.data["Sam Dat"]["Counts/Dt"])
axs[1].set_yscale('linear')

axs[2].set_ylabel('Norm')
axs[2].plot(ron.detector["Corr Depth"], ron.data["Sam Dat"]["Norm Cts"])
axs[2].set_yscale('linear')

axs[3].set_ylabel('Bgd Sub')
axs[3].plot(ron.detector["Corr Depth"], ron.data["Sam Dat"]["Corr Cts"])
axs[3].set_yscale('linear')

axs[4].set_ylabel('Atoms/cm2')
axs[4].plot(ron.detector["Corr Depth"], ron.data["Sam Dat"]["Atoms/cm2"])
axs[4].set_yscale('linear')

plt.show()
```
